```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.plot(np.random.randn(50).cumsum())
```

```python
points = np.arange(-5, 5, 0.01)
xs, ys = np.meshgrid(points, points)
z = np.sqrt(xs ** 2 + ys ** 2)
```

```python
plt.imshow(z, cmap=plt.cm.gray); plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```

```python
position = 0 
walk = [position]
steps = 100
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step 
    walk.append(position)
```

```python
plt.plot(walk[:10])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('Rone na Diya')
ax.set_xlabel('Bewaja')
ax.plot(np.random.randn(1070).cumsum())
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='r', alpha=0.654)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.87)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
 color='g', alpha=0.94321)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
