# Multiline plots

It's fun to plot a line, but it's even more fun when we can plot more than one line
on the same figure. This is really easy with Matplotlib as we can simply plot all
the lines that we want before calling show().

```python
x = range(1, 5)
plt.plot(x, [xi*1.5 for xi in x])
         
plt.plot(x, [xi*3.0 for xi in x])

plt.plot(x, [xi/3.0 for xi in x])
         
plt.show()
```

# Grid, axes, and labels


### Adding a grid

We can add a grid to the plot by calling the grid() function; it takes one parameter, a Boolean value, to enable (if True) or disable (if False) the grid:

```python
x = np.arange(1, 5)
plt.plot(x, x*1.75, x, x*4.2, x, x/2.0)
plt.grid(True)
plt.show()
```

# Handling axes

Let's take the first multiline plot.
Wouldn't it be better to have more spaces between lines and borders? We can achieve
this with the following code:

```python
x = np.arange(1, 5)
```

```python
plt.plot(x, x*1.5, x, x*3.0, x, x/3.0) 

plt.axis()
#If we execute axis() without parameters, it returns the actual axis limits.

#There are two ways to pass parameters to axis(): by a list of four values or by keyword arguments.
plt.axis([0, 5, -1, 13])

plt.show()
```

# Adding labels

To manage the axes dimensions is  another important piece of
information to add to a plot is the axes labels, since they usually specify what kind of
data we are plotting.

```python
plt.plot(8,4,'o')

plt.xlabel('X_Label')
plt.ylabel('Y_Label')
plt.title('Ploting Example')

plt.show()
```

```python
def create_random_walk():
    x = np.random.choice([-1,1],size=100, replace=True) # Sample with replacement from (-1, 1)
    return np.cumsum(x) # Return the cumulative sum of the elements

X = create_random_walk()
Y = create_random_walk()
Z = create_random_walk()

# Plotting functionality starts here
plt.plot(X)
plt.plot(Y)
plt.plot(Z)
plt.show()
```

```python
plt.plot(X, label="X")
plt.plot(Y, label="Y")
plt.plot(Z, label="Z")

#Adding Title
plt.title("Random Walk Example")

#Adding Labels
plt.xlabel("Number of Steps")
plt.ylabel("Accumulative Sum")
plt.show()
```

```python
plt.plot(X, label="X")
plt.plot(Y, label="Y")
plt.plot(Z, label="Z")

plt.title("Random Walk Example", fontsize=16, fontweight='bold')
plt.suptitle("Random Walk Suptitle", fontsize=10)
plt.xlabel("Number of Steps")
plt.ylabel("Accumulative Sum")
plt.show()

```

# Adding a legend

###### The last thing we need to see to complete a basic plot is a legend.

Legends are used to explain what each line means .Let's run with same example again.

```python
plt.plot(X, label="X")
plt.plot(Y, label="Y")
plt.plot(Z, label="Z")

#Adding Legend
plt.legend()

plt.title("Random Walk Example", fontsize=16, fontweight='bold')
plt.suptitle("Random Walk Suptitle", fontsize=10)
plt.xlabel("Number of Steps")
plt.ylabel("Accumulative Sum")
plt.show()
```

###### Here we can also change the location of the legend.

```python
plt.plot(X, label="abc")
plt.plot(Y, label="def")
plt.plot(Z, label="ghi")

plt.legend(loc='lower left')

plt.title("Random Walk Example", fontsize=17, fontweight='bold')
plt.suptitle("Random Walk Suptitle", fontsize=9)
plt.xlabel("Number of Steps" , fontweight = 25 , fontsize=10)
plt.ylabel("Accumulative Sum")
plt.show()
```
